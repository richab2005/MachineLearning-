```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('/content/drive/My Drive/ML/Data_SLR.xlsx')
df
```

```python
x= df.iloc[:, :-1].values
y=df.iloc[:, -1].values
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,color='blue')
plt.title('Years of Experience vs Salary')
plt.xlabel('Independent Variable = Years of Experience')
plt.ylabel('Dependent Variable = Salary')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2)
```

```python
print("Shape of x_train", x_train.shape)
print("Shape of y_train", y_train.shape)
print("Shape of x_test", x_test.shape)
print("Shape of y_test", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred= regressor.predict(x_test)
df1= pd.DataFrame({'Actual Values': y_test, 'Predicted Values': y_pred})
df1
```

```python
y_pred= regressor.predict(x_train)
df2= pd.DataFrame({'Actual Values': y_train, 'Predicted Values': y_pred})
df2
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

y_pred= regressor.predict(x_test)
mse=mean_squared_error(y_test, y_pred)
print("MSE= ",mse)

rmse= np.sqrt(mse)
print("RMSE= ",rmse)

mae=mean_absolute_error(y_test, y_pred)
print("MAE= ",mae)

r2= r2_score(y_test, y_pred)
print("R2= ",r2)
```

```python
y_pred= regressor.predict(x_train)
plt.scatter(x_train, y_train, color="green", label="Training data points(actual values)")
plt.scatter(x_train, y_pred, color="blue", label="Training data points(predicted values)")
plt.plot(x_train, y_pred, color='red')
plt.title("Salary vs Years of Experience(Training dataset)")
plt.xlabel("Years of Experinece")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
y_pred= regressor.predict(x_test)
plt.scatter(x_test, y_test, color="green", label="Testing data points(actual values)")
plt.scatter(x_test, y_pred, color="blue", label="Testing data points(predicted values)")
plt.plot(x_test, y_pred, color='red')
plt.title("Salary vs Years of Experience(Testing dataset)")
plt.xlabel("Years of Experinece")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
